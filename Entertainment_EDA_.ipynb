```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_basicinfo = pd.read_excel('/content/Entertainer - Basic Info.xlsx')
data_breakthroughinfo = pd.read_excel('/content/Entertainer - Breakthrough Info.xlsx')
data_lastworkinfo = pd.read_excel('/content/Entertainer - Last work Info.xlsx')
```

```python
# Breakthrough artists after 2000

breakthrough_after_2000 = data_breakthroughinfo[data_breakthroughinfo['Year of Breakthrough/#1 Hit/Award Nomination'] > 2000]
breakthrough_after_2000
```

```python
# Count of Breakthroughs year wise

grouped_by_breakthrough_year = data_breakthroughinfo.groupby('Year of Breakthrough/#1 Hit/Award Nomination').size()
grouped_by_breakthrough_year
```

```python
grouped_by_breakthrough_year.plot(kind='line', marker = 'o', figsize = (10, 6))
plt.xlabel('Year of Breakthrough')
plt.ylabel('Number of Breakthroughs')
plt.title('Number of Breakthroughs Year-wise')
plt.grid(True)
plt.show()
```

```python
# Histogram of award years

plt.figure(figsize = (12, 8))
plt.hist(data_breakthroughinfo['Year of First Oscar/Grammy/Emmy'].dropna() , bins = 10, edgecolor = 'r')
plt.xlabel('Year of Breakthrough')
plt.ylabel('Number of Artists')
plt.title('Histogram of Award Years')
plt.grid(True)
plt.show()
```

```python
# Heatmap of Award Year vs Breakthrough Year

pivot_table = data_breakthroughinfo.pivot_table(index = 'Year of Breakthrough/#1 Hit/Award Nomination', columns = 'Year of First Oscar/Grammy/Emmy', aggfunc = 'size', fill_value = 0)
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(pivot_table, cmap= 'YlGnBu', annot = True, fmt = 'd')
plt.title('Award Year vs Breakthrough Year Heatmap')
plt.xlabel('Year of First Oscar/Grammy/Emmy')
plt.ylabel('Year of Breakthrough/#1 Hit/Award Nomination')
plt.show()
```
