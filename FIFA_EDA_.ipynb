# 1. Historical Comparisons

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

Total Goals Scored by Year

```python
plt.figure(figsize= (10, 6))
sns.lineplot (data = world_cups, x = 'Year', y = 'GoalsScored', marker = 'o')
plt.title('Total Goals Scored in Each World Cup')
plt.xlabel('Year')
plt.ylabel('Total Goals')
plt.show()
```

Conclusion :

The total number of goals scored in world cup have increased over time, but with significant fluctuations. The most goals were scored in 1998 and 2014 tournaments, while the fewest were in the early editions and the 1950 world cup.  

Average Goals per match by year

```python
world_cups['AverageGoals'] = world_cups['GoalsScored'] / world_cups['MatchesPlayed']
plt.figure(figsize=(10,6))
sns.lineplot(data = world_cups, x = 'Year', y = 'AverageGoals', marker = 'o')
plt.title('Average Goals per Match by year')
plt.xlabel('Year')
plt.ylabel('Average Goals per Match')
plt.show()
```

Conclusion:

This plot reveals a significant decline in the average number of goals per match over the decades. This suggests a tactical evolution in the game , possibly with a focus on stronger defences and more strategic play.

Host Country Success

```python
host_success = world_cups.groupby('Country')['Winner'].count().sort_values(ascending= False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x = host_success.index , y = host_success.values, palette = 'viridis')
plt.title('Most Country Success in World Cups')
plt.xlabel('Country')
plt.ylabel('Number of Wins')
plt.xticks (rotation = 45)
plt.show()
```

Conclusion:

Several countries have won the worls cup on home soil, with Brazil, France, Germany , Italy and Mexico each having two home victories. However, some host countries have not won , suggesting that the home advantage doesn't guarantee a win.

Match Attendance over the years

```python
plt.figure(figsize= (10,6))
sns.lineplot(data = world_cup_matches , x = 'Year', y = 'Attendance', marker = 'o')
plt.title('Match Attendance over the years')
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.show()
```

Conclusion:

The world cup has seen a general increase in match attendance over the years, with a notable peak in 1990s . However, there's been fluctuation in attendance , suggesting varying levels of interest or accessibility depending on the host country or other factors.

Correlation Heatmap of World Cup Statistics

```python
# Convert 'Attendance' column to string type first if it's not already
world_cups['Attendance'] = world_cups['Attendance'].astype(str)

# Then, proceed with replacing '.' and converting to numeric
world_cups['Attendance'] = pd.to_numeric(world_cups['Attendance'].str.replace('.', ''), errors='coerce')


numeric_columns = ['Year', 'GoalsScored', 'QualifiedTeams', 'MatchesPlayed', 'Attendance', 'AverageGoals']
plt.figure(figsize= (10,8))
sns.heatmap(world_cups[numeric_columns].corr(), annot = True, cmap = 'coolwarm', linewidths=  0.5)
plt.title('Correlation Heatmap of World Cup Statistics')
plt.show()
```

Conclusion:

The world cup's popularity and scale have grown over time, evident in the increased goals, teams, matches and attendance. However, this growth has coincided with the a decrease in average goals per match , hinting at a possible shift towards more strategic , defensive play.

# Advanced Team Analysis

Top Scoring Teams

```python
top_teams = world_cup_players.groupby('Team Initials').size().sort_values(ascending = False).head(10)
plt.figure(figsize = (10,6) )
sns.barplot (x= top_teams.index, y = top_teams.values , palette = 'viridis')
plt.title('Top 10 teams by Number of Player appearances in world cup')
plt.xlabel('Team')
plt.ylabel('Number of Appearnces')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

Conclusion:

Brazil dominates with the highest number of player appearances in world cups, followed by Italy and Argentina. These top 3 teams significantly outpace other countries in terms of the cumulative world cup experience of their players.


Top Players by goals scored

```python
top_countries = world_cups.groupby('Country')['AverageGoals'].sum().sort_values(ascending = False).head(10)
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x = top_countries.index, y = top_countries.values, palette = 'viridis')
plt.title('Top 10 Countries by Goals scored in World Cups')
plt.xlabel('Country')
plt.ylabel('Total Goals')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

Conclusion:

Brazil and France are the top two goal-scoring countries in World Cup History , reflecting their consistent presence and attacking prowess.While European and South American teams dominate the top 10, Mexico's high ranking highlights the occasional success of teams from other regions.

Distribution of Match Outcomes (Win, Loss, Draw)

```python
world_cup_matches['Outcome'] = world_cup_matches.apply(
    lambda row: 'Home Win' if row['Home Team Goals'] > row['Away Team Goals'] else (
        'Away Win' if row['Home Team Goals'] < row['Away Team Goals'] else 'Draw'
    ),
    axis=1
)

outcome_counts = world_cup_matches['Outcome'].value_counts()
```

```python
plt.figure(figsize= (8,6))
sns.barplot(x = outcome_counts.index, y = outcome_counts.values, palette = 'viridis')
plt.title('Distribution of Match Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

Conclusion:

In this dataset the home teams enjoy a significant advantage , winning nearly twice as often as away teams. However, draws  remain a common outcome , suggesting a good degree of competitiveness in the matches.

# Additional Analysis

Dominance over Decades

```python
# convert year col to datetime format
world_cups['Year'] = pd.to_datetime(world_cups['Year'], format = '%Y')
```

```python
# extract decades
world_cups['Decade'] = world_cups['Year'].dt.year // 10 * 10
```

```python
# count the number of wins per team per decade

dominance_by_decade = world_cups.groupby(['Decade','Winner']).size().unstack(fill_value = 0)
```

```python
dominance_by_decade.plot(kind = 'bar', stacked = True, figsize = (10,6))
plt.title('Team Dominance over Decades in FIFA World Cup')
plt.xlabel('Decade')
plt.ylabel('Number of Wins')
plt.legend(title = 'Winning Teams', bbox_to_anchor = (1.05, 1), loc = 'upper left')
plt.tight_layout()
plt.show()
```

Conclusion:

This stacked bar chart provides a visual representation of the teams that dominated the world cup across different decades. It shows the repeated success of certain nations like Brazil, Germany , suggesting their long-term strength and consistent performance in the tournament.

Evolution of Strategies (Average goals per match by decade)

```python
avg_goals_per_decade = world_cups.groupby ('Decade')['AverageGoals'].mean()
```

```python
avg_goals_per_decade.plot(kind = 'line', marker = 'o', figsize = (10,6))
plt.title('Average Goals per Match by Decade')
plt.xlabel('Decade')
plt.ylabel('Average Goals per Match')
plt.grid(True)
plt.show()
```

Conclusion:

The graphy depicts a clear downward trend in the average number of goals scored per game in world cup matches over time. The peak occured in 1950s , with an average exceeding 4 goals per match. Since then, the average has steadily declined , reaching its lowest point in the 2000s and 2010s , with an average closer to 2.5 goals per match.

This suggests a shift in the style of play in world cup matches over decades , potentially towards a more defensive or tactical approach , resulting in fewer goals being scored.
