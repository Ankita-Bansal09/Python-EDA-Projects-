One Sample T Test

Let's suppose it was observed that the average purchase made by men of the age 18-25 years  was 10,000. Is it still the same?

```python
# write down alternate and null hypothesis
#create a subdata based on condition (requirement)
# we need to create a sample (random sampling)
# we perform hypothesis testing
# we conclude if its null or alternate hypothesis.
```

```python
# Null - average purchase made by men of age 18-25 years was 10,000.
# Alternate - average purchase made by men of age 18-25 years is not 10,000.
```

```python
new_data = data.loc[(data["Age"]== 1) & (data["Gender"]== 1)]
new_data
```

```python
sample_size = 1000
```

```python
sample = new_data.sample(sample_size , random_state= 0)
sample
```

```python
# one sample ttest

from scipy.stats import ttest_1samp
```

```python
t_stat, p_value= ttest_1samp(sample["Purchase"], 10000)
```

```python
p_value
```

```python
if p_value < 0.05:
  print("Null Hypothesis Rejected")
else:
  print("Null Hypothesis Accepted")
```

```python
#Two sample t test

```

Is the average purchase made by men and women of the age 18-25 same?

```python
# null- average purchase made by men and women of the age 18-25 same
# Alternate - average purchase made by men and women of the age 18-25 not same
```

```python
data_men = data.loc[(data["Age"]== 1) & (data["Gender"]== 1)]
data_women = data.loc[(data["Age"]== 1) & (data["Gender"]== 0)]
```

```python
data_men
```

```python
data_women
```

```python
sample_size = 500
```

```python
data_men_sample = data_men.sample(sample_size, random_state= 0)
data_women_sample = data_women.sample(sample_size, random_state= 0)
```

```python
data_men_sample
```

```python
data_women_sample
```

```python
# In a two sample T-Test we have to assume that the variance of the data is the same.
```

```python
print(data_men_sample["Purchase"].var())
print(data_women_sample["Purchase"].var())
```

```python
# Since they are not having exactly same, we have to assume that samples are having same variance.
```

```python
from scipy.stats import ttest_ind
```

```python
t_stat, p_value = ttest_ind(data_men_sample["Purchase"], data_women_sample["Purchase"], equal_var = True)
```

```python
p_value
```

```python
if p_value < 0.05:
  print("Alternate")
else:
  print("Null")
```

Mean of Product_category_1, Product_category_2 & Product_category_3 are the same.

Let's suppose it was observed that the percentage of women that spend more than 10,000 was 35 %. Is it still the same ?

```python
# Z-test for proportions
```

```python
#Null- percentage of women that spend more than 10,000 was 35 %.
#Alternate - percentage of women that spend more than 10,000 is not 35 %.

data_new = data.loc[(data["Purchase"]> 10000)]
```

```python
# num of women in this data

count = data_new["Gender"].value_counts()[0]
count
```

```python
# total num of obs in gender col

nobs = len(data_new["Gender"])
nobs
```

```python
# hypothesised value
p0 = 0.35
```

```python
from statsmodels.stats.proportion import proportions_ztest
z_stat, p_value = proportions_ztest(count= count , nobs= nobs , value= p0)
```

```python
if p_value < 0.05:
  print("Alternate")
else:
  print("Null")
```
