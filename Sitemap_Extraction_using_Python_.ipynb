# Extract Blog Pages URLs of Semrush Website

Locating the Sitemap : The sitemap needs to be located at https://www.semrush.com/sitemap.xml . Alternatively, a link to the sitemap in the robots.txt file,  can be found at https://www.semrush.com/robots.txt.

Sitemap Review: Review the nested sitemaps that may contain blog URLs. Nested sitemaps could appear as separate XML files, linked within the main sitemap. In the case of Semrush - https://www.semrush.com/blog/sitemap/ is the nested Sitemap which is inclusive of all the urls of Blogs.

Extract Blog URLs:
After identifying the blog-related sitemap, extract all relevant URLs.

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Function to extract URLs from a sitemap
def extract_sitemap_urls(sitemap_url):
    response = requests.get(sitemap_url)
    soup = BeautifulSoup(response.content, 'xml')  # Parse XML response
    urls = [url.text for url in soup.find_all('loc') if url.text is not None]  # Extract all URLs under the <loc> tag and check for None
    return urls
```

```python
sitemap_url = 'https://www.semrush.com/blog/sitemap/'
blog_urls = extract_sitemap_urls(sitemap_url)
```

```python
for url in blog_urls:
    print(url)
```
